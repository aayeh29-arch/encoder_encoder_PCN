```python
A = DensePCNLayer(1000, 0.00001, 'relu')
A.state = tf.Variable(tf.random.normal((10, 1000)))
B = DensePCNLayer(2000, 0.00001, 'relu')
A.next_layers=[B]
C = DensePCNLayer(3000, 0.00001, 'relu')
B.prev_layer = A
B.next_layers = [C]
C.prev_layer = B
b_out = B(A.state)
c_out = C(b_out)
```

```python
print(tf.math.reduce_std(B.state, axis=1))
print(tf.math.reduce_std(C.state, axis=1))
```

```python
for i in range(10):
      B.update_state()
      B.update_wts()
      B.update_b()
      print(tf.reduce_sum((B.predict_next() - B(A.predict_next()))**2)
            +tf.reduce_sum((C.predict_next() - C(B.predict_next()))**2)
            +tf.reduce_sum((A.predict_next()-B.predict_prev())**2)
            +tf.reduce_sum((B.predict_next()-C.predict_prev())**2))
```

```python

A = Conv2DPCNLayer(1, (3, 3), 0.0000001, 'relu')
A.state = tf.Variable(tf.random.normal((10, 100, 100, 1)))
B = Conv2DPCNLayer(5, (3, 3), 0.0000001, 'relu')
A.next_layers=[B]
C = Conv2DPCNLayer(9, (3, 3), 0.0000001, 'relu')
B.prev_layer = A
B.next_layers = [C]
C.prev_layer = B
b_out = B(A.state)
c_out = C(b_out)
```

```python
for i in range(10):
    B.update_state()
    B.update_wts()
    B.update_b()
    print(tf.reduce_sum((B.predict_next() - B(A.predict_next()))**2)
        +tf.reduce_sum((C.predict_next() - C(B.predict_next()))**2)
        +tf.reduce_sum((A.predict_next()-B.predict_prev())**2)
        +tf.reduce_sum((B.predict_next()-C.predict_prev())**2))
```

```python
A = DensePCNLayer(128, 0.00001)
C = DensePCNLayer(200, 0.00001)
A.state = tf.Variable(tf.random.normal((10, 1000, 128)))
B = TransformerPCNLayer(3, 128, 4, 0.00001, A, [C])
A.next_layers=[B]
C.prev_layer = B
b_out = B(A.state)
c_out = C(b_out)
```

```python
for i in range(10):
    for layer in B.get_layers():
        layer.update_state()
        layer.update_wts()
        layer.update_b()
    print(tf.reduce_sum((B.predict_next() - B(A.predict_next()))**2)
        +tf.reduce_sum((C.predict_next() - C(B.predict_next()))**2)
        +tf.reduce_sum((A.predict_next()-B.predict_prev())**2)
        +tf.reduce_sum((B.predict_next()-C.predict_prev())**2))
```
